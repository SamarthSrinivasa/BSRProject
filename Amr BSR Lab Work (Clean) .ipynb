```python
import pandas as pd
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

data = LAcity

def cluster_ethnicity_subregions(data, ethnicity_columns, epsilon, min_samples):
    
    ethnicity_data = data[ethnicity_columns]

    scaler = StandardScaler()
    ethnicity_data_scaled = scaler.fit_transform(ethnicity_data)

    dbscan = DBSCAN(eps=epsilon, min_samples=min_samples)

    clusters = dbscan.fit_predict(ethnicity_data_scaled)

    data['Cluster_Labels'] = clusters

    return data

ethnicity_columns = ['White', 'Afro_Amer', 'Native', 'Asian', 'Pac_Isl', 'Other_Race', 'Hispanic', 'Not_Hisp']
epsilon = 0.5
min_samples = 5

clustered_data = cluster_ethnicity_subregions(data, ethnicity_columns, epsilon, min_samples)

# The clustered_data DataFrame will now have an additional column 'Cluster_Labels' indicating the cluster for each subregion.
```
